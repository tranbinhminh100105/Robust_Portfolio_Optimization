```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
column_names = ['Date', 'Adj_Close', 'Close', 'High', 'Low', 'Open', 'Volume']
df = pd.read_csv('AAPL_history_data.csv', skiprows = 3, names = column_names)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df.head()
```

```python
df['Log_Return'] = np.log(df['Adj_Close'] / df['Adj_Close'].shift(1))
df.dropna(subset=['Log_Return'], inplace=True)
df.head(5)
```

```python
df['Daily_return'] = df['Adj_Close'].pct_change()
df['20Daily_return'] = df['Adj_Close'].pct_change(20)
df['20Daily_Votality'] = df['Daily_return'].rolling(20).std()
```
